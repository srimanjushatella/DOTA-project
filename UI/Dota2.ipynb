```python
# importing required packages
import matplotlib as plt
from matplotlib import style
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as mpt
```

```python
data = pd.read_csv('Actual.csv')
```

```python
data.head()
```

```python
%matplotlib inline
f,ax = mpt.subplots(figsize=(12, 12))
sb.heatmap(data.corr(), annot=True, linewidths=.5, fmt= '.1f', ax=ax)
```

```python
data.xp_per_min.plot(kind ='line', color = 'b', label = 'Experience', linewidth=2, alpha=1, grid = True, linestyle = '--')
data.gold_per_min.plot(kind = 'line', color = 'g', label = 'Gold collected', linewidth=2, alpha=1, grid = True, linestyle = ':')
mpt.legend(loc = 'upper right')
mpt.xlabel('X axis')
mpt.ylabel('Y axis')
mpt.title ('Line plot')
```

```python
data.plot(kind='scatter', x='xp_per_min', y='level',alpha = 0.5,color = 'red')
mpt.xlabel('Experience earner per minute')
mpt.ylabel('Level of Player')
mpt.title('Scatter plot')
```

```python
data.gold_per_min.plot (kind = 'hist', bins= 100, figsize= (15,15), color= 'lightblue')
mpt.xlabel('Gold Collected')
```
